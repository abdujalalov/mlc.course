```python
df = df.reset_index(drop=True)
df
```

```python
import seaborn as sns

# Create a line chart with confidence intervals
sns.lineplot(x='Year', y='Mobile cellular subscriptions (per 100 people)', data=df)

# Show the plot
plt.show()
```

```python
# Group the data by year and entity and calculate the mean value for each group
mean_values = df.groupby(['Year', 'Entity'])['Mobile cellular subscriptions (per 100 people)'].mean().reset_index()

# Create a line chart with confidence intervals
sns.lineplot(x='Year', y='Mobile cellular subscriptions (per 100 people)', hue='Entity', data=mean_values)

# Show the plot
plt.show()
```
